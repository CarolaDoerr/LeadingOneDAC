```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
```

```python
# plot total_evals
#policies = {'lo_50_123': ['random','fixed_one','optimal','optimal_discrete','dqn', 'dqn_gamma0.99'],
#            'lo_50_124': ['random','fixed_one','optimal','optimal_discrete','dqn'],
#            'lo_50_spread': ['random','fixed_one','optimal','optimal_discrete','dqn', 'dqn_gamma0.99'],
#           }
#policies = {'n50_power_of_2/k3': ['random','optimal_discrete','dqn'],
#            'n50_power_of_2/k4': ['random','optimal_discrete','dqn'],
#            'n50_power_of_2/k5': ['random','optimal_discrete','dqn']
#           }

# evenly spread
exps = [f"n50_evenly_spread/k{k}" for k in [3,4,5,6,7,8,10,15,20]]
exps = exps +  [f"n100_evenly_spread/k{k}" for k in [3,4,5,6,7,8,10,15,20]]
#exps = [f"n100_evenly_spread/k{k}" for k in [7,8,10,15,20]]
policies = {exp: ['random','optimal_discrete','dqn','dqn-1','dqn-2'] for exp in exps}
#policies = {exp: ['random','optimal_discrete','dqn'] for exp in exps}

# n=50, k=3,4,5; 3 settings
#exps = [f"{setting}/k{k}" for setting in ['n50_evenly_spread','n50_initial_segment','n50_power_of_2'] for k in [3,4,5]]
#policies = {exp: ['random','optimal_discrete','dqn'] for exp in exps}


plot_individual = False

medianprops = dict(linestyle='-', linewidth=1, color='red')
meanprops = dict(linestyle='dotted', linewidth=2, color='blue')
t = None
stats = []
nruns = 1
for exp in policies.keys():
    #print(f"\nExperiment: {exp}")
    stats.append(f"\nExperiment: {exp}")
    evals = {}        
    collected_policies = []
    for policy in policies[exp]:           
        data_fn = exp + "/" + policy + ".txt"
        if os.path.isfile(data_fn) is False:
            #print(f"{data_fn} is missing. Skipping...")
            continue
        with open(data_fn, "rt") as f:
            ls = [int(s[:-1]) for s in f.readlines()]            
            s = f"\t{policy:20s}: "                   
            if len(ls)>0:
                s += f"\t{np.mean(ls[:200]).round(0):6.0f} \t{np.quantile(ls[:200],[0.25,0.5,0.75]).round(0)}"
                s += f"\t{np.mean(ls[:1000]).round(0):6.0f} \t{np.quantile(ls[:1000],[0.25,0.5,0.75]).round(0)}"
                s += f"\t{np.mean(ls).round(0):6.0f} \t{np.quantile(ls,[0.25,0.5,0.75]).round(0)}"
                if len(ls) > nruns:
                    nruns = len(ls)
                if len(ls)==1:
                    ls = ls * nruns
                #print(s)             
                stats.append(s)
                evals[policy] = ls
                collected_policies.append(policy)
                          
    if plot_individual:
        plt.figure(figsize=(10,5));
        fig = plt.boxplot([vals for policy, vals in evals.items()], showmeans=True, meanline=True, medianprops=medianprops, meanprops=meanprops);
        ax = plt.gca();
        ax.set_xticklabels (labels=[f"{policy}" for policy in evals.keys()], rotation=0, ha='center');
        ax.set_ylabel("# evaluations")
        plt.title(f"{exp}") ;                 
    
    else:
        t1 = pd.DataFrame(evals)
        t1['exp'] = exp 
        t2 = pd.melt(t1, id_vars=['exp'], value_vars=collected_policies, var_name='policy', value_name='n_evals')        
        if t is None:
            t = t2
        else:
            t = pd.concat([t,t2], axis=0, ignore_index=True)
            
remove_random = False
use_boxplot = True
missing_val = 99999
log_scale = True
if plot_individual is False:
    t.loc[t.policy=='optimal', 'policy'] = 'optimal (unrestricted)'
    t.loc[t.policy=='optimal_discrete', 'policy'] = 'optimal' # rename optimal_discrete as optimal
    t['n'] = [int(exp.split('_')[0][1:]) for exp in t.exp]
    t['k'] = [int(exp.split('/')[-1].replace('k','')) for exp in t.exp]
    #t['setting'] = [exp.split('/')[0] for exp in t.exp]    
    t['setting'] = [exp.split('/')[0].replace('n50_','').replace('_',' ') for exp in t.exp] # setting name without n, for multiple-setting experiments only        
    print(t.setting.unique())
    if remove_random:
        t = t[t.policy!='random']
    for n in t.n.unique():        
        t1 = t[t.n==n]        
        max_val = t1[t1.n_evals!=missing_val].n_evals.max()
        #print(max_val)
        t1.loc[t1.n_evals==missing_val,'n_evals'] = max_val
        if use_boxplot:            
            #sns.set_context('poster', font_scale=1.2, rc={"lines.linewidth": 7})
            sns.set_context('paper', font_scale=3, rc={"lines.linewidth": 3})
            #sns.reset_orig()
            ax = sns.catplot(data=t1, x='k', y='n_evals', hue='policy', col='n', kind='box', legend=False)   
            ax.set_axis_labels("k","# iterations")
            if log_scale:
                ax.fig.get_axes()[0].set_yscale('log')
                #ax.fig.get_axes()[0].yticks([1,4,10,20,30])
            ax.fig.get_axes()[0].tick_params(axis='y', which='minor')
            #ax.yaxis.set_minor_formatter(FormatStrFormatter("%.1f"))
            #ax.fig.get_axes()[0].yaxis.set_major_formatter(FormatStrFormatter("%.1f"))
            #ax.set_titles(f"n={n}")            
            #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)            
            if n==50:
                ax.fig.set_figwidth(20)
                ax.fig.set_figheight(7) 
            else:
                ax.fig.set_figwidth(20)
                ax.fig.set_figheight(15) 
            ax.fig.set_figheight(7)                                                    
            plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2),fancybox=True, shadow=True, ncol=len(t.policy.unique()))
            plt.tight_layout()        
            #plt.title(f"n={n}");
            plt.savefig(f'plots/eval-n{n}_evenly_spread_boxplot.png')            
            #plt.savefig('plots/eval_n50_k456_three_settings_with_random_boxplot.png')            
            plt.show()
        else:            
            plt.figure(figsize=(12,5));            
            sns.lineplot(data=t1, x='k', y='n_evals', hue='policy', ci='sd', style='policy', markers=True, markersize=15, markeredgecolor=None, err_style='band', alpha=0.7)
            #plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
            ax = plt.gca();
            ax.set_xticks(t[t.n==n].k.unique());
            ax.set_xticklabels (labels=[str(k) for k in t1.k.unique()]);
            ax.set_ylabel("# evaluations")            
            plt.title(f"n={n}");
            plt.tight_layout()        
            plt.savefig(f'plots/eval-n{n}_evenly_spread_k7_20_3runs.png')
            plt.show()        

print("\n#evaluations across 200/1000/2000 runs. First column: average, remaining columns: 25th, 50th and 75th percentile")        
for s in stats:
    print(s)
```

```python
# plot learnt policy
import pickle
import json
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FormatStrFormatter

log_scale = True
#exps = ['lo_50_123','lo_50_124','lo_50_spread']
#exps = ['n50_power_of_2/k3','n50_power_of_2/k4','n50_power_of_2/k5']
#exps = [f"n50_evenly_spread/k{k}" for k in [3,4,5,6,7,8,10,15,20]]
#exps = exps +  [f"n100_evenly_spread/k{k}" for k in [3,4,5,6,7,8,10,15,20]]
exps = [f"n50_evenly_spread/k{k}" for k in [3,4,5]]
exps += [f"n50_initial_segment/k{k}" for k in [3,4,5]]
exps += [f"n50_power_of_2/k{k}" for k in [3,4,5]]

def get_optimal_policy(n):
    opt = pd.DataFrame({'fx':range(1,n+1), 'k': [n/(fx+1) for fx in range(1,n+1)]})
    return opt

def read_infos(fn, max_rows=100):
    with open(fn,"rb") as f:
        infos = pickle.load(f)
    t = None
    nrows = min(max_rows, len(infos))
    for info in infos[:nrows]:
        data = info['values']
        t1 = pd.DataFrame({'fx':[data['init_obj']] + data['log_fx'][:-1], 'k':data['log_k']})        
        if t is None:
            t = t1
        else:        
            t = pd.concat([t,t1],axis=0)
    t = t.drop_duplicates()
    return t

for exp in exps:
    n = int(exp.split('_')[0][1:])
    ddqn_fn = f"{exp}/dqn.pkl"
    if os.path.isfile(ddqn_fn) is False:
        print(f"{ddqn_fn} is missing. Skipping...")
        continue
    t = read_infos(ddqn_fn)
    t['policy'] = 'dqn'
    topt = get_optimal_policy(n)
    topt['policy'] = 'optimal (unrestricted)'
    topt_discrete = read_infos(f"{exp}/optimal_discrete.pkl")
    topt_discrete['policy'] = 'optimal'
    t = pd.concat([t,topt,topt_discrete],axis=0,ignore_index=True)
    sns.set_context('paper', font_scale=3, rc={"lines.linewidth": 3})
    plt.figure(figsize=(10,8));
    #ax = sns.lineplot(data=t, x="fx", y="k",ci=95, n_boot=100, hue='policy')
    ax = sns.lineplot(data=t, x="fx", y="k",ci=None, hue='policy', style='policy', markers=False, markersize=10, markeredgecolor=None, alpha=0.7)
    if log_scale:
        ax.set_yscale('log')
        plt.yticks([1,4,10,20,30])
        plt.tick_params(axis='y', which='minor')
        #ax.yaxis.set_minor_formatter(FormatStrFormatter("%.1f"))
        ax.yaxis.set_major_formatter(FormatStrFormatter("%.1f"))
    #ax.set_title(exp);
    n = exp.split('/')[0].split('_')[0].replace('n','')
    portfolio_setting = '_'.join(exp.split('/')[0].split('_')[1:])
    k = exp.split('/')[1].replace('k','')
    plt.title(f"{portfolio_setting} with n={n} and k={k}")
    ax.set_xlabel("f(x)")
    ax.legend().set_title('')
    #plt.savefig(f"{exp}/{exp.replace('/','_')}_dqn_policy.png")
    plt.savefig(f"{exp}/dqn_policy.png")
    plt.show()
```
