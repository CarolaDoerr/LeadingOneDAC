```python
import json
import pandas as pd
import numpy as np
import pickle
import gzip
from matplotlib import pyplot as plt
import seaborn as sb
import os
import pprint
import scipy.signal
import scipy.stats
import seaborn as sns
np.set_printoptions(suppress=True)
import matplotlib
import yaml
import glob
sns.set_style("white")
```

```python
# read all data
setting_filter = exp_filter = agent_filter = None
setting_filter = ['n150_evenly_spread', 'n200_evenly_spread']
#exp_filter = ['k3']
#agent_filter = ['ddqn','ddqn_batch8192']

# get all settings, exps and agents
settings = [s[:-1] for s in glob.glob('n*/')]
if setting_filter:
    settings = set(settings).intersection(set(setting_filter))
exps = []
for setting in settings:
    exps.extend([s.split('/')[-2] for s in glob.glob(setting + "/k*/")])
exps = list(set(exps))
agents = []
for setting in settings:
    for exp in exps:
        agents.extend([s.split('/')[-2] for s in glob.glob(setting + "/" + exp + "/ddqn*/")])
        seed_runs = glob.glob(setting + "/" + exp + "/ddqn*/seed_*/")
        if len(seed_runs)>0:
            agents.extend(['/'.join(s.split('/')[-3:-1]) for s in glob.glob(setting + "/" + exp + "/ddqn*/seed_*/")])
agents = list(set(agents))      

# read summary data of all chosen agents
all_data = {}
for setting in settings:
    for exp in exps:
        for agent in agents:
            data_dir = setting + '/' + exp + "/" + agent + "/"
            fn = data_dir + 'eval_infos.gzip'
            if os.path.isfile(fn) is False:
                continue
            print(f"Reading {data_dir}")
            # load agent's data
            data = []
            all_infos = []
            i = 0
            with gzip.open(fn, 'rb') as f:
                while True:
                    try:            
                        infos = [info[0]['values'] for info in pickle.load(f)]            
                        all_infos.append(infos)                   
                        ls = [rs['evals'] for rs in infos]
                        data.append([np.mean(ls), np.std(ls)])
                        print(str(i) + " evaluation iteration read", end='\r')
                        i += 1
                        #if i>=max_iters:
                        #    break
                    except EOFError:
                        break
            print("\n")            
            data = np.array(data)
            all_data[(setting, exp, agent)] = data
with open('ddqn_all_data_150_200.pkl','wb') as f:
    pickle.dump(all_data, f)
print("Finished")    
```

```python
def plot_progress(setting, exp, agent, max_steps=2_000_000, eval_interval=2_000):                    
    data = all_data[(setting, exp, agent)]        
    
    max_iters = int(max_steps / eval_interval)            
    data = data[:max_iters,:]

    # read ddqn experiment config
    data_dir = setting + '/' + exp + "/" + agent + "/"
    config_fn = data_dir + "/config.yml"
    with open(config_fn,'r') as f:
        params = yaml.safe_load(f)
    eval_interval = params['experiment']['eval_interval']

    # read optimal
    with open(setting + '/' + exp + "/optimal_discrete.txt", "rt") as f:
        ls = [int(s[:-1]) for s in f.readlines()]            
    mean_optimal = np.mean(ls)
    sd_optimal = np.std(ls)

    # where we hit the optimal
    close_to_theory = np.where(data[:,0] <= mean_optimal + .25*sd_optimal)[0]    

    # plot ddqn evaluation results              
    #sb.set_style('darkgrid')
    #sb.set_context('poster')
    fig = plt.figure(figsize=(15,8))
    # DAC
    plt.step(np.arange(1, len(data) + 1) * eval_interval, data[:,0], where='post', label='DDQN')
    #plt.plot(np.arange(1, len(data) + 1) * eval_interval, data[:,0], label='DDQN')
    u = data[:, 0] + data[:, 1]
    l = data[:, 0] - data[:, 1]
    plt.fill_between(np.arange(1, len(data) + 1) * eval_interval, u, l, alpha=0.1, step='post')

    # Theory
    plt.step(np.arange(1, len(data) + 1) * eval_interval, [mean_optimal for _ in range(len(data))],
             ls='--', label='optimal', where='post')
    u = mean_optimal + sd_optimal
    l = mean_optimal - sd_optimal
    plt.fill_between(np.arange(1, len(data) + 1) * eval_interval, u, l, alpha=0.1, step='post')

    # scatter points where we are close to theory
    plt.scatter(np.arange(1, len(data) + 1)[close_to_theory] * eval_interval, data[close_to_theory,0])

    plt.ylabel('# RLS iterations ')
    plt.xlabel('# train steps')
    #plt.title("Number of Evaluations during DDQN training phase: " + data_dir)
    #plt.title("Number of Evaluations during DDQN training phase " + data_dir)
    #plt.title()
    n = setting.split('_')[0].replace('n','')
    portfolio_setting = '_'.join(setting.split('_')[1:])
    k = exp.replace('k','')
    plt.title(f"{portfolio_setting} with n={n} and k={k}")
    
    plt.legend()
    #plt.xlim([0, 100_000])
    print(f"Save to {data_dir}/evals.png")
    plt.savefig(data_dir+'/evals.png')
    plt.show()
```

```python
def calculate_stats(setting, exp, agent, max_steps=2_000_000):
    data = all_data[(setting, exp, agent)]
    
    # read ddqn experiment config
    data_dir = setting + '/' + exp + "/" + agent + "/"
    config_fn = data_dir + "/config.yml"
    with open(config_fn,'r') as f:
        params = yaml.safe_load(f)
    eval_interval = params['experiment']['eval_interval']
    
    max_iters = int(max_steps / eval_interval)            
    data = data[:max_iters,:]

    # read optimal
    with open(setting + '/' + exp + "/optimal_discrete.txt", "rt") as f:
        ls = [int(s[:-1]) for s in f.readlines()]            
    mean_optimal = np.mean(ls)
    sd_optimal = np.std(ls)

    # hitting ratios
    close_to_theory = np.where(data[:,0] <= mean_optimal + .25*sd_optimal)[0]
    hit_percentage = np.round(len(close_to_theory)/data.shape[0]*100,0)  
    n_hits = len(close_to_theory)
    
    # smoothness
    sd_diff = np.std(np.diff(data[:,0])).round(2)
    avg_diff = np.mean(np.diff(data[:,0])).round(2)
    
    # lag one autocorrelation
    lagone = pd.Series(data[:,0]).autocorr(lag=1).round(2)    
    
    return {'hitting_ratio': np.round(hit_percentage/100,2), 'n_hits': n_hits, 'avg_diff': avg_diff, 'sd_diff': sd_diff, 'lagone': lagone}
    
t = pd.DataFrame(columns=['setting','setting_nice', 'n', 'k','hitting_ratio', 'n_hits', 'avg_diff', 'sd_diff', 'lagone', 'agent'])

# read data
with open('ddqn_all_data.pkl','rb') as f:
    all_data = pickle.load(f)
    
for setting in ['n50_evenly_spread','n50_power_of_2','n50_initial_segment']:
    for exp in ['k3','k4','k5']:
        for agent in ['ddqn']:

#for setting in ['n100_evenly_spread']:
#    for exp in ['k10','k15','k20']:   
#        for agent in ['ddqn']:

# gamma
#for setting in ['n50_evenly_spread','n100_evenly_spread']:
#    for exp in ['k3']:        
#        for agent in ['ddqn','ddqn_gamma0.99']:
            
# batch size
#for setting in ['n50_evenly_spread']:

#for setting in ['n150_evenly_spread','n200_evenly_spread']:
#    for exp in ['k3','k4','k5']:   
#        for agent in ['ddqn/seed_0','ddqn/seed_1','ddqn/seed_2']:
        
            #agent = 'ddqn'        
            s = '_'.join(setting.split('_')[1:])
            n = int(setting.split('_')[0][1:])
            k = int(exp.replace('k',''))
            stats = calculate_stats(setting, exp, agent)
            #setting_nice = f"n={n}, seed={agent.split('_')[1]}" # with agent seed and without setting
            #setting_nice = f"n={n}" # without agent seed and without setting
            #setting_nice = f"n={n}, {' '.join(setting.split('_')[1:])}, seed={agent.split('_')[1]}" #with everything
            setting_nice = f"n={n}, {' '.join(setting.split('_')[1:])}" #with setting but no agent seed
            rs = {'setting':[setting],'k':[k], 'agent':agent, 'setting_nice':setting_nice, 'n':n}        
            for key, val in stats.items():
                rs[key] = val
            row = pd.DataFrame(rs)
            t = pd.concat([t,row],axis=0, ignore_index=True)                   
display(t.sort_values(by='k'))   
```
